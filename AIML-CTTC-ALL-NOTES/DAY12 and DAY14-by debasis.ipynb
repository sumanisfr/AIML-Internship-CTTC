```python
#CREATION OF IP/OP:-
# we take a variable ip that will store all input values except the target one for here tv radio and newspaper
ip = avt.drop('Sales',axis = 1)
ip.head()
```

```python
op = avt.Sales
op.head()
```

<H1> DAY 14 STARTS</H1>

```python
#Train Test Split:-
#splitting 100% of the data to training data to training and testing data.
# We need to give 80 percent data to machine
from sklearn.model_selection import train_test_split
```

```python
#Input Training data = xtrain
#Output Training data  = ytrain
#Input Testing data  = xtest
#Output Testing data  = ytest
#train_size = 0.8
#or
#test_size = 0.2
xtrain,xtest,ytrain,ytest = train_test_split(ip,op,test_size = 0.8 '''random_state = 5''') 
#it executes untill  5 times 5th one will be final
```

```python
print(xtrain.head())
```

```python
print(xtest.head())
```

```python
print(ytrain.head())
```

```python

print(ytest.head())
```

```python
#standardising the data using standarscaler library 
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
```

```python
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

```python
xtest
```

```python
#linear Regression  aim is to find the best fit line
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
# machine learning done
```

```python
# now its testing time
pred = lr.predict(xtest)
pred
```

```python
# to find the accuracy of the model we need to find the 
#MSE =  mean square error 
#and 
# r2score
```

```python
from sklearn.metrics  import mean_squared_error
error = mean_squared_error(pred,ytest) #orginal value and tested value
error 
```

```python
# r2Score:-it is used to find the accuracy of the model 
from sklearn.metrics import r2_score
acc = r2_score(pred,ytest)
acc
```

```python
# to plot the best plot line we need to plot the lm plot
# lm plot stands for linear model plot
```

```python
df = pd.DataFrame({'Y_Test':list(ytest),'Prediction':list(pred)})
```

```python
df
```

```python
sns.lmplot(x = 'Y_Test',y = 'Prediction',data = df)
plt.show()
```
