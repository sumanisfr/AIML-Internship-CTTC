```python
# INCORRECT (causes SyntaxError):
# xtrain,xtest,ytrain,ytest = train_test_split(ip,op,test_size = 0.8,'''random_state = 5''')

# CORRECT version:
xtrain, xtest, ytrain, ytest = train_test_split(ip, op, test_size=0.2, random_state=5)

# Alternative correct versions:
# Using train_size instead
# xtrain, xtest, ytrain, ytest = train_test_split(ip, op, train_size=0.8, random_state=5)

# With proper spacing
# xtrain, xtest, ytrain, ytest = train_test_split(ip, op, test_size=0.2, random_state=5)

# Complete working example:
from sklearn.model_selection import train_test_split
import numpy as np

# Sample data (replace with your actual data)
ip = np.random.rand(100, 4)  # 100 samples, 4 features
op = np.random.randint(0, 2, 100)  # 100 binary target values

# Correct train-test split
xtrain, xtest, ytrain, ytest = train_test_split(ip, op, test_size=0.2, random_state=5)

print(f"Training data shape: {xtrain.shape}")
print(f"Testing data shape: {xtest.shape}")
print(f"Training labels shape: {ytrain.shape}")
print(f"Testing labels shape: {ytest.shape}")
```

```python
# print(xtrain.head())
# print(xtrain.head())  # This won't work with NumPy arrays
# Option 1: Array slicing (recommended)
print(xtrain[:5])  # Shows first 5 rows

# Option 2: Convert to DataFrame first
print(pd.DataFrame(xtrain).head())

# Option 3: Check shape and first few elements
print(f"Shape: {xtrain.shape}")
print(f"First 5 rows:\n{xtrain[:5]}")
```

```python
#standardising the data using standarscaler library 
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
```

```python
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

```python
xtest
```

```python
#linear Regression  aim is to find the best fit line
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
# machine learning done
```

```python
# now its testing time
pred = lr.predict(xtest)
pred
```

```python
# to find the accuracy of the model we need to find the 
#MSE =  mean square error 
#and 
# r2score
```

```python
from sklearn.metrics  import mean_squared_error
error = mean_squared_error(pred,ytest) #orginal value and tested value
error 
```

```python
# r2Score:-it is used to find the accuracy of the model 
from sklearn.metrics import r2_score
acc = r2_score(pred,ytest)
acc
```

```python
# to plot the best plot line we need to plot the lm plot
# lm plot stands for linear model plot
```

```python
df = pd.DataFrame({'Y_Test':list(ytest),'Prediction':list(pred)})
```
