### Creating Shapes:-

```python
# drawing a line:-
# syntax :- cv2.line(start_cordinate(x,y),end_cordinate(x1,y1),color(rgb),Thinkness)
```

```python
#creating an array of black color:-
r = np.full((50,50),0)   #((arraysize),value of color)
g = np.full((50,50),0)   
b = np.full((50,50),0)
rgb1 = np.dstack((r,g,b))
plt.imshow(rgb1)
# plt.axis(False)
plt.axis('off')
plt.show()
```

```python
#shortcut for black color array create
p = np.zeros((500,500,3))
plt.imshow(p)
plt.show()
```

```python
cv2.line(p,(0,500),(500,0),(255,255,255),4)
plt.imshow(p)
plt.show()
```

```python
l = cv2.line(p,(0,0),(500,500),(255,255,255),4)
plt.imshow(l)
plt.show()
```

```python
r = cv2.line(l,(0,250),(500,250),(255,255,255),4)
plt.imshow(r)
plt.show()
```

```python
r = cv2.line(l,(250,500),(250,0),(45,55,255),4)
plt.imshow(r)
plt.show()
```

```python
# cv2.circle(center(x,y),radius,color,thickness)
c = np.zeros((500,500,3))
cv2.circle(c,(250,250),100,(255,0,0),4)
plt.imshow(c)
plt.show()
```

```python
c = np.zeros((500,500,3))
c1 = cv2.circle(c,(250,250),100,(255,0,0),4)
plt.imshow(c)
plt.show()
```

```python
c2 = cv2.circle(c1,(250,250),50,(255,0,0),4)
plt.imshow(c2)
plt.show()
```

```python
# draw some square :-
# cv2.rectangle(1stdiagonalcordinate(x,y),2nddiagonalcordinate(x1,y1),color,thickness)
cv2.rectangle(c1,(400,100),(100,400),(0,255,0),4)
plt.imshow(c1)
plt.show()
```

```python
s1 = cv2.rectangle(p,(400,100),(100,400),(0,255,0),4)
plt.imshow(s1)
plt.show()
```

```python
z = np.zeros((500,500,3))
s1 = cv2.rectangle(z,(400,100),(100,400),(0,255,0),4)
plt.imshow(s1)
plt.show()
```

```python
# converting BGR to grey :-
gray = cv2.cvtColor(army,cv2.COLOR_RGB2GRAY)
cv2.imshow('Gray',gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Converting BGR to RGB.
rgb = cv2.cvtColor(army,cv2.COLOR_BGR2RGB)
# cv2.imshow('rgb',rgb)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
plt.imshow(rgb)
plt.show()
```

```python
# Detection of Edges:-
edge = cv2.Canny(rgb,30,200)
plt.imshow(edge,cmap='gray')
plt.show()
```
