```python
x = np.array([3,7,5,8,11,15])
y = np.array([2,6,10,11,13,9])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# Single Layer Perceptron:-
model = Sequential()
# input & hidden layer :-
model.add(Dense(1,input_dim =1,activation='linear'))
#output layer :-
model.add(Dense(1,activation='linear'))
model.compile(Adam(learning_rate=0.1),loss='mse')
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=20,verbose=0)
```

```python
# Applying Multi-layer-perceptron Mode:-
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
# Creating the Model:-
model=Sequential()
# input and hidden layer :-
model.add(Dense(30,input_dim=3,activation='linear'))

# hidden layer:-
model.add(Dense(15,activation='linear'))
model.add(Dense(25,activation='linear'))

# output layer :-
model.add(Dense(1,activation='linear'))
model.compile(Adam(learning_rate=0.01),loss='mse')
```
