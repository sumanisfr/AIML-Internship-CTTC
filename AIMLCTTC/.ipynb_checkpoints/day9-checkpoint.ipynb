## New csv file(Auto-mpg)

```python
#importing the libraries :-
import numpy as np
import pandas as pd
```

```python
# Reading the dataset 
auto = pd.read_csv(r"C:\Users\suman\OneDrive\Documents\Desktop\AIML-CTTC-ALL-NOTES\auto-mpg.csv")
auto
```

```python
auto.shape
```

```python
# to check null value
auto.isnull().sum()
```

```python
# to check datatype
auto.dtypes
```

```python
# find the unique values :-
for i in auto.columns:
    print(f"{i}:/n {auto[i].unique()}\n")       #horsepower is now string so check unique
```

```python
# to find the total no. of missing values (?) in dataset.
sum(auto['mpg']=='?')
```

```python
sum(auto['horsepower']=='?')
```

```python
# to check all in dataset so use for loop
for i in auto.columns:
    print(f"{i} : {sum(auto[i] == '?')}")
```

```python
des = auto.describe(include='all')
des
```

```python
# replacing the '?' with top repeated value.
auto['horsepower'].replace('?',des['horsepower'][2])
```

```python
for i in auto.columns:
    print(f"{i} : {sum(auto[i] == '?')}")
```

```python
# to permanetly change 
auto['horsepower'].replace('?',des['horsepower'][2],inplace=True)
```

```python
for i in auto.columns:
    print(f"{i} : {sum(auto[i] == '?')}")
```

```python
auto.dtypes
```

```python
# to change the datatypes of 'horsepower' explicitly.
auto.horsepower = auto.horsepower.astype('float')
auto.dtypes
```

```python
auto.describe(include='all')
```

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60]
y = [2000,2001,2002,2003,2004,2005]
plt.plot(x,y)  #to the graph using x and y data.
plt.show()     # to plot the graph
```

```python
x = [45,89,23,67,45,67,27]
y = [2000,2001,2002,2003,2004,2005,2006]
plt.title('Line plot',c='red') # gives a heading to the graph
plt.plot(x,y,marker='o')
plt.xlabel('Datas') # Gives a name to x-axis
plt.ylabel('Year') # gives a name to y-axis
plt.show()
```

```python
x = ['Team A','Team B','Team C']
y = [288,567,345]
colours = ['yellow','green','orange']
plt.bar(x,y,color = colours)  # plot a bar graph
plt.title('Bar Plot')
plt.xlabel('Teams')
plt.ylabel('No. of Members')
plt.show()
```
