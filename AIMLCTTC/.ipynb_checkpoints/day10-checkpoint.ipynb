### Scatter plot :- used to Display the Relationship between 2 vareiables.

```python
x  = [3,46,75,88,22,99,34,56,92]
y = [1,4,28,24,73,99,45,32,87]
plt.scatter(x,y, marker ='s',s=200,color='gold')
plt.title('Scatter plot')
plt.show()
```

### Histogram :- Used to display the frequence of the data distribution.

```python
age = [13,43,22,44,87,76,42,63,56,36,20,48,52,81,77,88]
bins = [10,20,30,40,50,60,70,80,90]    #for age grp
plt.hist(age,bins,histtype='bar',color='gold',edgecolor='k',rwidth=0.7)
plt.title('Histogram')
plt.xlabel('interval')
plt.ylabel('Ages')
plt.show()
```

### pie plot /pie chart:- Used to display data as a properation of any whole

```python
activities = ['playing','sleeping','eating','roaming','studying']
slices = [6,12,2,3,1]
cl = ['blue','red','gold','green','pink']
plt.pie(slices,labels =activities,colors = cl,startangle=45,shadow=True, explode=(0.1,0,0,0,0),autopct='%1.1f%%')
plt.title("pie-Chart")
plt.show()
```

# Data Visualization :-
# Distribution Plot:-Used to show the distribution of any single variable.
<pre> combines the histogram and kernak density estimate plot(KDE)</pre>

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
# Data Visualization :-
# Distribution Plot:-
sns.distplot(iris.SL)
plt.show()
```

### Displot :- shows the distribution of a single variable histogram.

```python
sns.displot(iris.SL)
plt.show()
```

```python
sns.displot(iris.SL,fill=False)
plt.show()
```

```python
sns.displot(iris.SL,fill=False,hatch = '*',color ="gold")
plt.show()
```

#### KDE plot :- Kernal Density Estimaten plot is used to display the frequncey distribution of any single varible using curved line

```python
sns.kdeplot(iris.SW,fill=True,hatch='*',color ="gold")
plt.show()
```

```python
sns.kdeplot(iris.SL,fill=True,hatch='*',color ="indigo")
plt.show()
```

```python
sns.distplot(iris.PL[iris.Class=='Iris-setosa'],label='Iris-setosa')
sns.distplot(iris.PL[iris.Class=='Iris-versicolor'],label='Iris-versicolor')
sns.distplot(iris.PL[iris.Class=='Iris-virginica'],label = 'Iris-virginica')
plt.legend()
plt.show()
```

```python
sns.histplot(iris.PL[iris.Class=='Iris-setosa'],label='Iris-setosa')
sns.histplot(iris.PL[iris.Class=='Iris-versicolor'],label='Iris-versicolor')
sns.histplot(iris.PL[iris.Class=='Iris-virginica'],label = 'Iris-virginica')
plt.legend()
plt.show()
```

```python
sns.displot(iris.PL[iris.Class=='Iris-setosa'],label='Iris-setosa')
sns.displot(iris.PL[iris.Class=='Iris-versicolor'],label='Iris-versicolor')
sns.displot(iris.PL[iris.Class=='Iris-virginica'],label = 'Iris-virginica')
plt.legend()
plt.show()
```

```python
sns.kdeplot(iris.PL[iris.Class=='Iris-setosa'],label='Iris-setosa')
sns.kdeplot(iris.PL[iris.Class=='Iris-versicolor'],label='Iris-versicolor')
sns.kdeplot(iris.PL[iris.Class=='Iris-virginica'],label = 'Iris-virginica')
plt.legend()
plt.show()
```

### Box plot :- Used to Display the statistical data of the variable.

```python
sns.boxplot(x=iris.Class,y=iris.SW)
plt.show()
```

```python
sns.boxplot(x=iris.Class,y=iris.SW,palette='cool')
plt.show()
```

### Violin plot :- used to display the distribution , median and density of the datas across the categories

```python
sns.violinplot(x=iris.Class,y=iris.SW,palette='hot')
plt.show()
```

```python
sns.scatterplot(x=iris.Class,y=iris.SW,palette='hot')
plt.show()
```

### Swarm plot :- Used to display the statistical datas of each categories using all points.


```python
sns.swarmplot(x=iris.Class,y=iris.SL,palette='hot')
plt.show()
```

```python
sns.swarmplot(x=iris.Class,y=iris.SL,palette='colorblind')
plt.show()
```

#### Bar plot :- Used to Display the Summary statistic(mean) of anumeric variable along with  Categorical data 
<pre>It automatically represent the cofidence interval(CI) 
CI:- Range of values that acually are true </pre>

```python
sns.barplot(x=iris.Class,y=iris.SL,palette='cool')
plt.show()
```

```python
plt.figure(figsize = (20,25))
sns.countplot(x=iris.SL,palette='cool')
plt.show()
```

```python
sns.countplot(iris.SL,palette='cool')
plt.show()
```

#### Scatter plot :- Used to display the relationship between 2 numerical value

```python
sns.scatterplot(x=iris.SL,y=iris.PL,palette='cool',hue=iris.Class)
plt.show()
```

### Joint plot :- used to display bivariate scatter plot of each variable.

```python
sns.jointplot(data=iris)
plt.show()
```

```python
sns.jointplot(data=iris,kind='kde')
plt.show()
```

#### Pair plot :- create a grid of plots that display the pairwise relationship between multiple numerical columns in the dataset.

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris,hue='Class')
plt.show()
```
