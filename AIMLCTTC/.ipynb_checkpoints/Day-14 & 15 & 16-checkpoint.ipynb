## SVM(support vector machine)

## SVM :- support vector machine

#### Its 2 two types. 
* SVR:- Support vector regressor
* SVC:- Support vector classifer :- to find the best fit hyperplane/line thats separates the categories in the best way 

<pre> kernel function :-   # kernel function Aims to converts 1D to 2D, 2D to 3D.
1.linear -> 1D to 2D
2. poly -> 2D to 3D
3. sigmoid -> 2D to 3D
4. rbf:-(Radial Basis Function) -> 1D to 2D to 3D </pre>

<pre> Gamma -> coffiecient of the kernel function. </pre>

<pre> C = Regularization parameter (this restrict our model to be overfitted) 
    Range of C :-
                1. (0.01 - 1) - small(noisy data, wants to avoid underfitting )
                2. (  1 -10 ) - Medium(good trainig starting point)
                3. (10-10000) - High(data is clean,helps to minimize underfitting)</pre>

```python
from sklearn.svm import SVC 
sv = SVC( kernel = 'rbf',gamma = 0.01, C =1000)       
sv.fit(xtrain,ytrain)
```

```python
pred2 = sv.predict(xtest)
pred2
```

```python
# from sklearn.matrics import accuracy_score,recall_score
# acc = accuracy_score(pred2,ytest)
# rec = recall_score(pred2,ytest)
# print(acc)
# print(rec)
from sklearn.metrics import recall_score,accuracy_score
rs = recall_score(pred2,ytest)
ac = accuracy_score(pred2,ytest)
print(f"Accuracy:",ac)
print(f"Recall:",rs)
```

```python
# confusion matrices
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix
cm = confusion_matrix(pred2,ytest)
cm2 = ConfusionMatrixDisplay(cm)
cm2.plot()
plt.show()
```

# Naive Bayes :- It works on probability.
<pre> Naive Bayes uses Sayes theorem to predict the categories of a new data point.</pre>

```python
# Bayes Theorem :- [P(A/B) = (P(B/A).P(A))/P(B)]
```

```python
# 3 coins  --> 2 to the power 3 = 8 
# s= {HHH,HHT,HTH,HTT,TTT,TTH,THT,THH}
# P(A) = {TTT,TTH,HTT,THT}    #occuring TT
# P(B/A) = {TTT,TTH}   # Starting with TT
# P(A/B) = P(AnB).P(A)/P(B)
# P(B/A) = P(AnB).P(B)/P(A)
```

<pre> There are 3 kind of Bayes theorem :-
1. Guassian Bayes Theorem :- Data is numerical as well as categoriecal datas
2. Multinomial :- only for text datas 
3. Bernoulli :-Only for categoriecal datas /pre>

## k-Means Clustering:- It belongs to Exclussive UnSupervised ML algorithm
<pre> Clustering :- means grouping up of the datas
        by using Euclidean distance formula 
    ----> Mean of all data point near to the particular cluster </pre>

```python
iris = pd.read_csv(r"C:\Users\suman\OneDrive\Documents\Desktop\AIML-CTTC-ALL-NOTES\iris\iris.data",header=None)
iris
```

```python
iris.columns = ['Sepal_Length','Sepal_width','Petal_Lenght','Petal_width','Class']
iris
```

```python
# to find the unique values of the columns at a times.
for i in iris.columns:
    print(f"{i}:\n {iris[i].unique()}\n")
```

```python
ip = iris.drop('Class',axis=1)
ip.head()
```

```python
# To apply the K-mean-Clustering Algorithm:-
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3)
# km 
km.fit(ip)
```

```python
km.fit(ip)
```

```python
# prediction:- 
km_pred = km.predict(ip)
km_pred
```

```python
centroids = km.cluster_centers_
centroids
```

```python
iris['Prediction'] = km_pred
iris
```

```python
sns.scatterplot(x =iris.Sepal_Length,y = iris.Petal_Lenght )
plt.show()
```

```python
sns.scatterplot(x =iris.Sepal_Length,y = iris.Petal_Lenght , c = km_pred)
plt.show()
```

```python
# Ploting along the centroid values:-
sns.scatterplot(x =iris.Sepal_Length,y = iris.Petal_Lenght , c = km_pred)
sns.scatterplot (x=centroids[:,0] , y =centroids[:,2], s=300,marker='*',color='k')
plt.show()
```
